```python
%matplotlib inline

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
from sklearn.decomposition import PCA, TruncatedSVD
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import scale

from sklearn.model_selection import train_test_split
from sklearn import decomposition
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor

import seaborn as sns
np.random.seed(0)
```

```python
df1 = pd.read_csv("Google_Trends_concat.csv")
df1.head(2)
```

```python
df1 = df1.set_index(["Year_week"])
df1.head()
```

```python
target = df1["label"]
```

```python
data = df1.drop("label", axis=1)
feature_names = data.columns
data.head()
```

```python
X = data.values
```

```python
y = target
```

```python
X.shape, y.shape
```

```python
# Set up a train-test split, for now we'll do it randomly
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 25) # 20% hold for testing
```

```python
rreg = RandomForestRegressor(max_depth = 2, random_state = 0)
```

```python
rreg.fit(X_train, y_train)
```

```python
y_pred = rreg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```
