```python
# This codes breaks *very easily* and is not well-written.  Please re-factor when you get a chance
# Make sure the number of columns in flu_season_compact_w_yr_df is the same as the number in the denominator on
# the second to last line.  Also, make sure the final number in the line beginning with 'next_season' is one less than 
# the number of columns in flu_season_compact_w_yr_df

# Also, consider iterating over the DataFrame in a better way.  This only is reasonable because the DataFrame is 
# small 

# The 'Season' column *must* be the last column in the DataFrame,
next_season_means = []
for ix in range(len(flu_compact_w_yr_df)):
    next_season = flu_compact_w_yr_df.iat[ix,33]
    state = flu_compact_w_yr_df.index[ix]
    source1_df = flu_compact_w_yr_df.loc[state]
    source_df = source1_df.loc[source1_df['Season'] == next_season]
    next_season_mean = source_df.mean(axis = 1) - next_season / 34
    next_season_means.append(next_season_mean)
```
